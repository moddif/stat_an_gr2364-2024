## Основные теоретические положения

Плотность нормального распределения:

$p_{\overline{\xi}} \left(\overline{x}\right) = \cfrac{1}{\left(2\pi\right)^{n/2} \sqrt{\det\Sigma}} \cdot \exp \left(-\cfrac{1}{2}\left(\overline{x}-\mu\right)^{\intercal}\Sigma^{-1}\left(\overline{x}-\mu\right)\right)$

Математическое ожидание и ковариационная матрица:

$\mathbb{E}\left(A\overline{\xi}+\overline{b}\right) = A\cdot\mathbb{E}\overline{\xi}+\overline{b}$

$\mathrm{Var}\left(A\overline{\xi}+\overline{b}\right)=A\cdot\mathrm{Var}\left(\overline{\xi}\right)\cdot A^{\intercal}$

Стандартное нормальное распределение:

$\mathbb{E} \begin{pmatrix} \xi_{st} \\ \eta_{st} \end{pmatrix} = \begin{pmatrix} 0 \\ 0 \end{pmatrix}$

$\mathrm{Var} \begin{pmatrix} \xi_{st} \\ \eta_{st} \end{pmatrix} = \begin{pmatrix}1 & 0 \\ 0 & 1\end{pmatrix}$

### Задание 1

Исходная функция распределения:

$p_{\xi, \eta} (x, y) = C \cdot \exp \left(-\cfrac{1}{2} \left(7x^2+8xy+13y^2+6x-18y+12\right)\right)$

Преобразуем полином таким образом, что бы чётко прослеживалось математическое ожидание и обратная ковариационная матрица:

$7x^2+8xy+13y^2+6x-18y+12 = 7\left(x^2+\cfrac{8}{7}xy+\cfrac{16}{49}y^2\right)-7\cdot\cfrac{16}{49}y^2+13y^2+6x-18y+12 = 7\left(x+\cfrac{4}{7}y\right)^2+\cfrac{75}{7}y^2+6x-18y+12 =$

$= 7\left(x+\cfrac{4}{7}y\right)^2+6\left(x+\cfrac{4}{7}y\right)-6\cdot\cfrac{4}{7}y-18y+\cfrac{75}{7}y^2+12 = 7\left(\left(x+\cfrac{4}{7}y\right)^2+\cfrac{6}{7}\left(x+\cfrac{4}{7}y\right)+\cfrac{9}{49}\right)-7\cdot\cfrac{9}{49}+\cfrac{75}{7}y^2+\cfrac{150}{7}y+12 =$

$= 7\left(x+\cfrac{4}{7}y+\cfrac{3}{7}\right)^2+\cfrac{75}{7}y^2+\cfrac{150}{7}y+\cfrac{75}{7} = 7\left(x+\cfrac{4}{7}y+\cfrac{3}{7}\right)^2+\cfrac{75}{7}\left(y-1\right)^2$

Запомним данное выражение (оно потребуется во втором задании):

$\star = 7\left(x+\cfrac{4}{7}y+\cfrac{3}{7}\right)^2+\cfrac{75}{7}\left(y-1\right)^2$

Продолжим:

$7\left(x+\cfrac{4}{7}y+\cfrac{3}{7}\right)^2+\cfrac{75}{7}\left(y-1\right)^2 = 7\left(\left(x+1\right)+\cfrac{4}{7}\left(y-1\right)\right)^2+\cfrac{75}{7}\left(y-1\right)^2 = 7\left(x+1\right)^2+8\left(x+1\right)\left(y-1\right)+13\left(y-1\right)^2$

Математическое ожидание:

$\mathbb{E} \begin{pmatrix}\xi \\ \eta\end{pmatrix} = \begin{pmatrix}-1 \\ 1\end{pmatrix}$

Ковариационная матрица:

$\Sigma^{-1} = \begin{pmatrix}7 & 4 \\ 4 & 13\end{pmatrix}$

$\Sigma = \cfrac{1}{75}\begin{pmatrix}13 & -4 \\ -4 & 7\end{pmatrix}$

Дисперсии:

$\mathbb{D} \xi = \cfrac{13}{75}$

$\mathbb{D} \eta = \cfrac{7}{75}$

Ковариация и корреляция:

$\mathrm{cov} (\xi, \eta) = -\cfrac{4}{75}$

$\rho (\xi, \eta) = \cfrac{\mathrm{cov} (\xi, \eta)}{\sqrt{\mathbb{D}\xi\mathbb{D}\eta}} = -\cfrac{4}{\sqrt{91}}$

Константа:

$\det \Sigma = \cfrac{1}{75}, n=2$

$C = \cfrac{1}{\left(2\pi\right)^{n/2}\sqrt{\det\Sigma}} = \cfrac{1}{2\pi\sqrt{\cfrac{1}{75}}} = \cfrac{5\sqrt{3}}{2\pi}$

$p_{\xi, \eta} (x, y) = \cfrac{5\sqrt{3}}{2\pi} \cdot \exp \left(-\cfrac{1}{2} \left(7x^2+8xy+13y^2+6x-18y+12\right)\right)$

График:

```python
import numpy as np
import matplotlib.pyplot as plt
from math import pi, e

x = np.arange(-2, 0, 0.1)
y = np.arange(0, 2, 0.1)
xgrid, ygrid = np.meshgrid(x, y)
zgrid = 5 * 3 ** 0.5 / (2 * pi) * e ** (-0.5 * (7 * xgrid ** 2 + 8 * xgrid * ygrid + 13 * ygrid ** 2 + 6 * xgrid - 18 * ygrid + 12))
fig = plt.figure()
ax = plt.axes(projection="3d")
ax.plot_wireframe(xgrid, ygrid, zgrid)
plt.show()
```

### Задание 2

Воспользуемся выражением из первого задания:

$\star = 7\left(x+\cfrac{4}{7}y+\cfrac{3}{7}\right)^2+\cfrac{75}{7}\left(y-1\right)^2 = \left(\sqrt{7}x+\cfrac{4\sqrt{7}}{7}y+\cfrac{3\sqrt{7}}{7}\right)^2+\left(\cfrac{5\sqrt{3}}{\sqrt{7}}y-\cfrac{5\sqrt{3}}{\sqrt{7}}\right)^2$

Преобразование имеет вид:

$\begin{pmatrix} \xi_{st} \\ \eta_{st} \end{pmatrix} = A\begin{pmatrix} \xi \\ \eta \end{pmatrix} + \overline{b}$

Найдём матрицу $A$ и вектор $\overline{b}$:

$A = \begin{pmatrix} \sqrt{7} & \cfrac{4\sqrt{7}}{7} \\ 0 & \cfrac{5\sqrt{3}}{\sqrt{7}} \end{pmatrix}, \overline{b} = \begin{pmatrix} \cfrac{3\sqrt{7}}{7} \\ -\cfrac{5\sqrt{3}}{\sqrt{7}} \end{pmatrix}$

Подставим и посчитаем:

$\begin{pmatrix} \xi_{st} \\ \eta_{st} \end{pmatrix} = \begin{pmatrix} \sqrt{7} & \cfrac{4\sqrt{7}}{7} \\ 0 & \cfrac{5\sqrt{3}}{\sqrt{7}} \end{pmatrix}\begin{pmatrix} \xi \\ \eta \end{pmatrix} + \begin{pmatrix} \cfrac{3\sqrt{7}}{7} \\ -\cfrac{5\sqrt{3}}{\sqrt{7}} \end{pmatrix} = \begin{pmatrix} \sqrt{7}\xi+\cfrac{4\sqrt{7}}{7}\eta+\cfrac{3\sqrt{7}}{7} \\ \cfrac{5\sqrt{3}}{\sqrt{7}}\eta-\cfrac{5\sqrt{3}}{\sqrt{7}} \end{pmatrix}$

Проверим полученное выражение:

$\mathbb{E} \begin{pmatrix} \xi_{st} \\ \eta_{st} \end{pmatrix} = \begin{pmatrix} \sqrt{7} & \cfrac{4\sqrt{7}}{7} \\ 0 & \cfrac{5\sqrt{3}}{\sqrt{7}} \end{pmatrix}\begin{pmatrix} -1 \\ 1 \end{pmatrix} + \begin{pmatrix} \cfrac{3\sqrt{7}}{7} \\ -\cfrac{5\sqrt{3}}{\sqrt{7}} \end{pmatrix} = \begin{pmatrix} -\sqrt{7}+\cfrac{4\sqrt{7}}{7}+\cfrac{3\sqrt{7}}{7} \\ \cfrac{5\sqrt{3}}{\sqrt{7}}-\cfrac{5\sqrt{3}}{\sqrt{7}} \end{pmatrix} = \begin{pmatrix} 0 \\ 0 \end{pmatrix}$

$\mathrm{Var} \begin{pmatrix} \xi_{st} \\ \eta_{st} \end{pmatrix} = A\Sigma A^{\intercal} = \cfrac{1}{75}\begin{pmatrix} \sqrt{7} & \cfrac{4\sqrt{7}}{7} \\ 0 & \cfrac{5\sqrt{3}}{\sqrt{7}} \end{pmatrix}\begin{pmatrix}13 & -4 \\ -4 & 7\end{pmatrix}\begin{pmatrix} \sqrt{7} & 0 \\ \cfrac{4\sqrt{7}}{7} & \cfrac{5\sqrt{3}}{\sqrt{7}} \end{pmatrix} = \cfrac{1}{75}\begin{pmatrix}\cfrac{75}{\sqrt{7}} & 0 \\ -\cfrac{20\sqrt{3}}{\sqrt{7}} & 5\sqrt{21}\end{pmatrix}\begin{pmatrix} \sqrt{7} & 0 \\ \cfrac{4\sqrt{7}}{7} & \cfrac{5\sqrt{3}}{\sqrt{7}} \end{pmatrix} = \cfrac{1}{75}\begin{pmatrix}75 & 0 \\ 0 & 75\end{pmatrix} = \begin{pmatrix}1 & 0 \\ 0 & 1\end{pmatrix}$

Математическое ожидание имеет нулевой вектор, а ковариационная матрица является единичной, следовательно, найденное нормальное распределение является стандартным.
